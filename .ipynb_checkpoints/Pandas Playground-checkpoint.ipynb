<h2>Create from a Dictionary</h2>

```python
# The important thing to note here is that you need to access the DataFrame type to create a dataframe from a dictionary.
tempDict = { 'col1': [1, 2, 3], 'col2': [4, 5, 6], 'col3': [7, 8, 9] }
dictdf = pd.DataFrame.from_dict(tempDict)
dictdf
```

<h1>Read</h1>

<h2>Filtering Columns</h2>

```python
# You could filter by using `df.State`, however, this method will only work on single word features.
df['State']
```

```python
# You can filter multiple features at the same time.
df[['State', 'International plan']]
```

```python
# This will give you a list of unique values.
df['Area code'].unique()
```

<h2>Filtering Rows</h2>

```python
# You can create a list based on a condition.
df['International plan']=='No'
```

```python
# Printing the entire dataframe with values that meet the condition.
df[df['International plan']=='No']
```

```python
# Multiple conditions; Notice the parenthesis.
df[(df['International plan']=='No') & (df['Churn']==False)]
```

<h2>iLoc</h2>

```python
df.iloc[12]
```

```python
# First parameter is the row, second is the column; iloc is zero indexed.
df.iloc[12, 2]
```

```python
# You can get a range of values.
df.iloc[22:33, :3]
```

```python
# This is a common pattern you will see.
# The dependent variable is tranditionally the last column. This will extract all of the rows of the dependent columns
# The first parameter is all rows
# The second parameter is up to the excluding the last column (independent variable).
df.iloc[:, :-1]
```

<h2>loc</h2>

```python
# When we copy in a csv, the index is automatically created as integer values.
# When using loc, we need something else to be the index.
state = df.copy()
state.set_index('State', inplace=True)
state
```

```python
# We are using the loc method to get all of the rows that are OH. 
# We could do this with a conditional, however, this might be slightly cleaner.
state.loc['OH']
```

<h1>Update</h1>

```python
# This will give you information on how many values are null in each feature.
df.isnull().sum()
```

```python
# This will drop all of the rows that have a null value
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# We can drop a column. You can add an inplace parameter to update the df object.
df.drop('Area code', axis=1)
```
